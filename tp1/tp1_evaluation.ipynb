# DÃ©veloppement:

#### importation:

```python
import random
import matplotlib.pyplot as plt
```

#### calcule de la moyenne empirique:

```python
def Xn(t,n):
    x=0
    for j in t:
        x=x+j
    return x/n
```

#### calcule de la variance empirique:

```python
def On(t,n,xn):
    tita=0
    for i in t:
        tita=tita+(i-xn)**2
    return tita/n
```

#### crÃ©ation des valeurs alÃ©ratoires avec la distribution uniforme:

```python
def creat_alea_table(n):
    t=[]
    for i in range(n+1):
        t.append(random.uniform(0, 1)) #les valeurs de i compris entre 0 et 1.
    return t
```

```python
n=[1000,10000,100000,1000000] #les Ã©chantillons
titatable=[]
xn=[]
for i in n:
    t=creat_alea_table(i)
    xn.append(Xn(t,i))
    titatable.append(On(t,i,Xn(t,i)))
```

#### ReprÃ©sentation graphique:

```python
plt.plot(n,titatable,'r*')
plt.plot(n,xn,'b*')
plt.xlabel("n:les valeurs alÃ©ratoires")
plt.ylabel("moyenne/ variance")
plt.show()
for i in range(len(n)):
    print ('Nombre des variables aleatoire Xi est= {} ,la moyenne= {} ,la variance= {}.'.format(n[i],xn[i],titatable[i]))
```

# InterprÃ©tation:

##### comparaison la rÃ©sultat thÃ©orique avec les rÃ©sultats pratique:

#### conclusion:

#### D'aprÃ¨s le tableau comparatif, on constate que le valeur de la moyenne empirique et la varience empirique sont presque egal l'esperence et la varience theorique quelque soit le nombre de variable aleatoire (N)

```python
def creat_tabe(n):
    x=[]
    t=np.random.uniform(0, 1,n)
    for i in t:
        if i<0.3:
            x.append(0)
        elif i<0.5:
            x.append(2)
        else:
            x.append(6)
    return x
```

```python
xn=[]
tita=[]
for i in n:
    t=creat_tabe(i)
    x=Xn(t,i)
    xn.append(x)
    tita.append(On(t,i,x))
for i in range(len(n)):
    print ('Nombre des variables aleatoire Xi est= {} ,la moyenne= {} ,la variance= {}.'.format(n[i],xn[i],tita[i]))
```
